```python
import numpy as np
import os
import time
```

Lire le contenu des fichiers texte_{i} (entrer le numero du texte)

```python
all_files = os.listdir("textes")
text = [file for file in all_files if file.endswith(".txt")]

messages = {}
result = {}

for file in text:
    with open("textes/" + file, "r") as f:
        messages[file] = f.read()
        
print(result)
print(messages)
```

Construction du dictionnaire à partir des symboles contenus dans le message. (Code fourni sur moodle)

```python
for file in text:
    Message = messages[file]
    dictsymb =[Message[0]]
    dictbin = ["{:b}".format(0)]
    nbsymboles = 1
    for i in range(1,len(Message)):
        if Message[i] not in dictsymb:
            dictsymb += [Message[i]]
            dictbin += ["{:b}".format(nbsymboles)] 
            nbsymboles +=1
            
    longueurOriginale = np.ceil(np.log2(nbsymboles))*len(Message)

    result[file] = {
        "dictsymb": dictsymb,
        "dictbin": dictbin,
        "nbsymboles": nbsymboles,
        "longueurOriginale": longueurOriginale
    }

for file, data in result.items():
    print(f"File: {file}")
    for key, value in data.items():
        print(f"{key}: {value}")
    print()
```

Ajustement des codes binaires selon le nombre total de symboles en ajoutant des zéros. Tri des symboles. Affichage du dictionnaire initial. (Code fourni sur moodle)

```python
for file in result:
    for i in range(result[file]["nbsymboles"]):
        nbsymboles = result[file]["nbsymboles"]
        dictsymb = result[file]["dictsymb"]
        dictbin = result[file]["dictbin"]

        dictbin[i] = "{:b}".format(i).zfill(int(np.ceil(np.log2(nbsymboles))))

    dictsymb.sort()
    dictionnaire = np.transpose([dictsymb, dictbin])
    result[file]["dictionnaire"] = dictionnaire

for file, data in result.items():
    print(f"File: {file}")
    print(result[file]["dictionnaire"])
```

Codage du message (Code fourni sur moodle)

```python
for file in result:
    i=0
    MessageCode = []
    longueur = 0
    nbsymboles = result[file]["nbsymboles"]
    dictsymb = result[file]["dictsymb"]
    dictbin = result[file]["dictbin"]

    tempsDebut = time.time()

    Message = messages[file]
    while i < len(Message):
        precsouschaine = Message[i] #sous-chaine qui sera codé
        souschaine = Message[i] #sous-chaine qui sera codé + 1 caractère (pour le dictionnaire)
        
        #Cherche la plus grande sous-chaine. On ajoute un caractère au fur et à mesure.
        while souschaine in dictsymb and i < len(Message):
            i += 1
            precsouschaine = souschaine
            if i < len(Message):  #Si on a pas atteint la fin du message
                souschaine += Message[i]  

        #Codage de la plus grande sous-chaine à l'aide du dictionnaire  
        codebinaire = [dictbin[dictsymb.index(precsouschaine)]]
        MessageCode += codebinaire
        longueur += len(codebinaire[0]) 
        #Ajout de la sous-chaine codé + symbole suivant dans le dictionnaire.
        if i < len(Message):
            dictsymb += [souschaine]
            dictbin += ["{:b}".format(nbsymboles)] 
            nbsymboles +=1
        
        #Ajout de 1 bit si requis
        if np.ceil(np.log2(nbsymboles)) > len(MessageCode[-1]):
            for j in range(nbsymboles):
                dictbin[j] = "{:b}".format(j).zfill(int(np.ceil(np.log2(nbsymboles))))

    tempsFin = time.time()
    
    result[file]["MessageCode"] = MessageCode
    result[file]["longueur"] = longueur
    result[file]["temps"] = round(tempsFin - tempsDebut, 5)

```

```python
for file, data in result.items():
    print(f"File: {file}")
    print(result[file]["MessageCode"])
```
