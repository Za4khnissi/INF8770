```python
import numpy as np
from PIL import Image
import binascii
```

Lire le contenu des images et convertir chaque pixel vers une seule representation hexadecimale (chat-gpt)

```python
image_number = input("Quel image voulez-vous lire ? (entre 1 et 5) ")
file_name = f"images/image_{image_number}.png"
image = Image.open(file_name)

width, height = image.size
Message = ""

for x in range(width):
    for y in range(height):
        coordinate = (x, y)
        pixel_value = image.getpixel(coordinate)
        hex_value = binascii.hexlify(bytes(pixel_value)).decode('utf-8')
        Message += hex_value
        
print(Message)
```

Ajustement des codes binaires selon le nombre total de symboles (chaque symbole est une chaine de six characteres representant le code hexadecimal) en ajoutant des zros. Tri des symboles. Affichage du dictionnaire initial.

```python
dictsymb =[Message[0:6]]
dictbin = ["{:b}".format(0)]
nbsymboles = 1
for i in range(6,len(Message), 6):
    if Message[i:i+6] not in dictsymb:
        dictsymb += [Message[i:i+6]]
        dictbin += ["{:b}".format(nbsymboles)]
        nbsymboles +=1
        
longueurOriginale = np.ceil(np.log2(nbsymboles))*len(Message)

print("Dictionary:")
for i in range(len(dictsymb)):
    print(f"{dictsymb[i]}: {dictbin[i]}")
print(f"Original Length: {longueurOriginale}")
```

Ajustement des codes binaires selon le nombre total de symboles en ajoutant des zros. Tri des symboles. Affichage du dictionnaire initial. (Code fourni sur moodle)

```python
for i in range(nbsymboles):
    dictbin[i] = "{:b}".format(i).zfill(int(np.ceil(np.log2(nbsymboles))))
        
dictsymb.sort()
dictionnaire = np.transpose([dictsymb,dictbin])
print(dictionnaire)
```

Codage du message (code fourni sur moodle)

```python
i=0
MessageCode = []
longueur = 0
while i < len(Message):
    precsouschaine = Message[i:i+6] #sous-chaine qui sera cod
    souschaine = Message[i:i+6] #sous-chaine qui sera cod + 1 caractre (pour le dictionnaire)
    
    #Cherche la plus grande sous-chaine. On ajoute un caractre au fur et  mesure.
    while souschaine in dictsymb and i < len(Message):
        i += 6
        precsouschaine = souschaine
        if i < len(Message):  #Si on a pas atteint la fin du message
            souschaine += Message[i:i+6]  

    #Codage de la plus grande sous-chaine  l'aide du dictionnaire  
    codebinaire = [dictbin[dictsymb.index(precsouschaine)]]
    MessageCode += codebinaire
    longueur += len(codebinaire[0]) 
    #Ajout de la sous-chaine cod + symbole suivant dans le dictionnaire.
    if i < len(Message):
        dictsymb += [souschaine]
        dictbin += ["{:b}".format(nbsymboles)] 
        nbsymboles +=1
    
    #Ajout de 1 bit si requis
    if np.ceil(np.log2(nbsymboles)) > len(MessageCode[-1]):
        for j in range(nbsymboles):
            dictbin[j] = "{:b}".format(j).zfill(int(np.ceil(np.log2(nbsymboles))))
```

Affichage du message cod

```python
print("Message cod : ",MessageCode)
```

Affichage du dictionnaire final

```python
dictionnaire = np.transpose([dictsymb,dictbin])
print(dictionnaire) 
```

```python
print("Longueur = {0}".format(longueur))
print("Longueur originale = {0}".format(longueurOriginale))
print("Ratio = {0}".format(longueur/longueurOriginale))
```
